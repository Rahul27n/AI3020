```python
import pandas as pd
```

```python
nodes = pd.read_csv('Points.csv')
```

```python
nodes
```

```python
edges = pd.read_csv('Links.csv')
```

```python
edges
```

```python
node_ids = list(nodes['IP Address of Node'])
```

```python
d = {}

for id in node_ids:
    d[id] = edges.stack().value_counts()[id]
```

```python
degrees = list(d.values())
```

```python
nodes['Total_Degree'] = degrees
```

```python
nodes
```

```python
nodes.to_csv('Points_With_Degree.csv',index=False)
```
